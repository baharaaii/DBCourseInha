```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('film_rental.db')
cursor = conn.cursor()

# Create Tables:

# LANGUAGE Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS LANGUAGE (
        Id INTEGER PRIMARY KEY,
        Language_name TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        UNIQUE(Language_name)
    );
""")

# DIRECTOR Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS DIRECTOR (
        Id INTEGER PRIMARY KEY,
        Last_name TEXT NOT NULL,
        First_name TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# COUNTRY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS COUNTRY (
        Id INTEGER PRIMARY KEY,
        Country_name TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# CATEGORY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS CATEGORY (
        Id INTEGER PRIMARY KEY,
        Category_name TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# STORE Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS STORE (
        Id INTEGER PRIMARY KEY,
        Address_Id TEXT NOT NULL,
        Phone_number TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# STAFF Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS STAFF (
        Id INTEGER PRIMARY KEY,
        Address_Id TEXT NOT NULL,
        Last_name TEXT NOT NULL,
        First_name TEXT NOT NULL,
        Email TEXT NOT NULL,
        Store_Id INTEGER,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# CUSTOMER Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS CUSTOMER (
        Id INTEGER PRIMARY KEY,
        Last_name TEXT NOT NULL,
        First_name TEXT NOT NULL,
        Email TEXT NOT NULL,
        Address_Id TEXT,
        Status INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# CITY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS CITY (
        Id INTEGER PRIMARY KEY,
        City_name TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# ADDRESS Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS ADDRESS (
        Id INTEGER PRIMARY KEY,
        Postal_code INTEGER NOT NULL,
        City_Id INTEGER NOT NULL,
        District TEXT NOT NULL,
        Street TEXT NOT NULL,
        Apartment_number TEXT NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT
    );
""")

# FILM Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS FILM (
        Id INTEGER PRIMARY KEY,
        Title TEXT NOT NULL,
        Release_year DATE NOT NULL,
        Description TEXT,
        Price REAL NOT NULL,
        Deposit REAL NOT NULL,
        Rating REAL,
        Length INTEGER NOT NULL,
        Language_Id INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        UNIQUE(Title)
    );
""")

# FILM_CATEGORY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS FILM_CATEGORY (
        Id INTEGER PRIMARY KEY,
        Film_id INTEGER NOT NULL,
        Category_id INTEGER,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Category_id) REFERENCES CATEGORY(Id),
        FOREIGN KEY(Film_id) REFERENCES FILM(Id)
    );
""")

# FILM_DIRECTOR Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS FILM_DIRECTOR (
        Id INTEGER PRIMARY KEY,
        Film_id INTEGER NOT NULL,
        Director_id INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Director_id) REFERENCES DIRECTOR(Id),
        FOREIGN KEY(Film_id) REFERENCES FILM(Id)
    );
""")

# FILM_COUNTRY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS FILM_COUNTRY (
        Id INTEGER PRIMARY KEY,
        Film_Id INTEGER NOT NULL,
        Country_Id INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Country_Id) REFERENCES COUNTRY(Id),
        FOREIGN KEY(Film_Id) REFERENCES FILM(Id)
    );
""")

# PRODUCT Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS PRODUCT (
        Id INTEGER PRIMARY KEY,
        Film_Id INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Film_Id) REFERENCES FILM(Id)
    );
""")

# INVENTORY Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS INVENTORY (
        Id INTEGER PRIMARY KEY,
        Store_Id INTEGER NOT NULL,
        Product_Id INTEGER NOT NULL,
        Availability_status BOOLEAN INTEGER NOT NULL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Product_Id) REFERENCES PRODUCT(Id),
        FOREIGN KEY(Store_Id) REFERENCES STORE(Id)
    );
""")

# RENTAL Table
cursor.execute("""
    CREATE TABLE IF NOT EXISTS RENTAL (
        Id INTEGER PRIMARY KEY,
        Inventory_Id INTEGER NOT NULL,
        Staff_Id INTEGER NOT NULL,
        Customer_Id INTEGER NOT NULL,
        Rental_date DATE NOT NULL,
        Return_date DATE,
        Amount REAL,
        CreatedOn DATETIME NOT NULL,
        CreatedBy TEXT,
        ChangedOn DATETIME NOT NULL,
        ChangedBy TEXT,
        FOREIGN KEY(Inventory_Id) REFERENCES INVENTORY(Id),
        FOREIGN KEY(Customer_Id) REFERENCES CUSTOMER(Id),
        FOREIGN KEY(Staff_Id) REFERENCES STAFF(Id)
    );
""")

# Commit changes and close connection
conn.commit()
conn.close()
```

```python
import sqlite3

#Abdulkhak's queries
# Connect to SQLite database
conn = sqlite3.connect('film_rental.db')
cursor = conn.cursor()

# Intermediate Queries:

# 1. List the films along with their categories
cursor.execute("""
    SELECT Film.Title, GROUP_CONCAT(Category.Category_name, ', ') AS Categories
    FROM Film
    JOIN Film_Category ON Film.Id = Film_Category.Film_id
    JOIN Category ON Film_Category.Category_id = Category.Id
    GROUP BY Film.Id;
""")
result1 = cursor.fetchall()
print("\nIntermediate Query 1: Films and Categories")
for row in result1:
    print(row)

# 2. Retrieve the total number of rentals for each customer
cursor.execute("""
    SELECT Customer.Id, Customer.First_name, Customer.Last_name, COUNT(Rental.Id) AS TotalRentals
    FROM Customer
    LEFT JOIN Rental ON Customer.Id = Rental.Customer_Id
    GROUP BY Customer.Id;
""")
result2 = cursor.fetchall()
print("\nIntermediate Query 2: Total Rentals for Each Customer")
for row in result2:
    print(row)

# 3. Find the staff members who have rented films
cursor.execute("""
    SELECT DISTINCT Staff.Id, Staff.First_name, Staff.Last_name
    FROM Staff
    JOIN Rental ON Staff.Id = Rental.Staff_Id;
""")
result3 = cursor.fetchall()
print("\nIntermediate Query 3: Staff Members Who Have Rented Films")
for row in result3:
    print(row)

# Advanced Queries:

# 4. Calculate the average rating for films in each language using CTE
cursor.execute("""
    WITH AvgRatingPerLanguage AS (
        SELECT Language.Id, Language.Language_name, Film.Rating,
                AVG(Film.Rating) OVER (PARTITION BY Film.Language_Id) AS AvgRating
        FROM Film
        JOIN Language ON Film.Language_Id = Language.Id
    )
    SELECT DISTINCT Language_name, AvgRating
    FROM AvgRatingPerLanguage;
""")
result4 = cursor.fetchall()
print("\nAdvanced Query 4: Average Rating for Films in Each Language")
for row in result4:
    print(row)

# 5. Identify the customers who have rented films in multiple categories using CTE
cursor.execute("""
    WITH CustomerCategoryCount AS (
        SELECT Customer.Id, COUNT(DISTINCT Film_Category.Category_id) AS CategoryCount
        FROM Customer
        JOIN Rental ON Customer.Id = Rental.Customer_Id
        JOIN Inventory ON Rental.Inventory_Id = Inventory.Id
        JOIN Film_Category ON Inventory.Product_Id = Film_Category.Film_id
        GROUP BY Customer.Id
    )
    SELECT Customer.Id, Customer.First_name, Customer.Last_name
    FROM Customer
    JOIN CustomerCategoryCount ON Customer.Id = CustomerCategoryCount.Id
    WHERE CustomerCategoryCount.CategoryCount > 1;
""")
result5 = cursor.fetchall()
print("\nAdvanced Query 5: Customers Who Have Rented Films in Multiple Categories")
for row in result5:
    print(row)

# 6. Calculate Cumulative Rental Revenue for Each Film Over Time
cursor.execute("""
    WITH CumulativeRevenuePerFilm AS (
        SELECT
            Inventory.Product_Id AS Film_Id,
            Rental.Rental_date,
            Rental.Amount,
            SUM(Rental.Amount) OVER (PARTITION BY Inventory.Product_Id ORDER BY Rental.Rental_date) AS CumulativeRevenue
        FROM Rental
        JOIN Inventory ON Rental.Inventory_Id = Inventory.Id
    )
    SELECT
        CumulativeRevenuePerFilm.Film_Id,
        Film.Title,
        CumulativeRevenuePerFilm.Rental_date,
        CumulativeRevenuePerFilm.Amount,
        CumulativeRevenuePerFilm.CumulativeRevenue
    FROM CumulativeRevenuePerFilm
    JOIN Film ON CumulativeRevenuePerFilm.Film_Id = Film.Id;
""")
result_cumulative_revenue = cursor.fetchall()

# Print the result of the cumulative revenue query
print("\nAdvanced Query: Cumulative Rental Revenue for Each Film Over Time")
for row in result_cumulative_revenue:
    print(row)

# Close the connection
conn.close()
```
